```python
import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
## Function to load images and masks from multiple directories
def load_images_and_masks(image_dirs, mask_dirs, img_size):
    images = []
    masks = []
    for image_dir, mask_dir in zip(image_dirs, mask_dirs):
        image_paths = sorted([os.path.join(image_dir, f) for f in os.listdir(image_dir) if f.endswith('.jpg')])
        mask_paths = sorted([os.path.join(mask_dir, f) for f in os.listdir(mask_dir) if f.endswith('.png')])

        for img_path, mask_path in zip(image_paths, mask_paths):
            img = load_img(img_path, target_size=img_size)
            img = img_to_array(img) / 255.0  # Normalize to [0, 1]
            mask = load_img(mask_path, target_size=img_size, color_mode="grayscale")
            mask = img_to_array(mask)
            mask = mask / np.max(mask)  # Normalize to [0, 1]
            mask = (mask > 0.5).astype(np.uint8)  # Binarize mask
            images.append(img)
            masks.append(mask)
    return np.array(images), np.array(masks)

# Define image directories
image_dirs = [
    "C:/Users/Abhishek/Downloads/archive (6)/TrainingImages/TrainingImages/OriginalImages",
    "C:/Users/Abhishek/Downloads/archive (6)/TrainingImages/TrainingImages/rotate_real/90_degrees",
    "C:/Users/Abhishek/Downloads/archive (6)/TrainingImages/TrainingImages/rotate_real/180_degrees",
    "C:/Users/Abhishek/Downloads/archive (6)/TrainingImages/TrainingImages/rotate_real/270_degrees"
]

# Define mask directories
mask_dirs = [
    "C:/Users/Abhishek/Downloads/archive (6)/TrainingImages/TrainingImages/EnumMasks/png_0_255",
    "C:/Users/Abhishek/Downloads/archive (6)/TrainingImages/TrainingImages/Rotated/90_degrees",
    "C:/Users/Abhishek/Downloads/archive (6)/TrainingImages/TrainingImages/Rotated/180_degrees",
    "C:/Users/Abhishek/Downloads/archive (6)/TrainingImages/TrainingImages/Rotated/270_degrees"
]

# Define image size
img_size = (128, 128)

# Load images and masks
images, masks = load_images_and_masks(image_dirs, mask_dirs, img_size)
```

```python
# Function to define the U-Net model
def get_unet_model(img_size, num_classes):
    inputs = keras.Input(shape=img_size + (3,))
    x = layers.Conv2D(32, 3, strides=2, padding="same")(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
    previous_block_activation = x  # Set aside residual

    for filters in [64, 128, 256]:
        x = layers.Activation("relu")(x)
        x = layers.SeparableConv2D(filters, 3, padding="same")(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)
        x = layers.SeparableConv2D(filters, 3, padding="same")(x)
        x = layers.BatchNormalization()(x)
        x = layers.MaxPooling2D(3, strides=2, padding="same")(x)

        # Project residual
        residual = layers.Conv2D(filters, 1, strides=2, padding="same")(previous_block_activation)
        x = layers.add([x, residual])
        previous_block_activation = x  # Set aside next residual

    for filters in [256, 128, 64, 32]:
        x = layers.Activation("relu")(x)
        x = layers.Conv2DTranspose(filters, 3, padding="same")(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)
        x = layers.Conv2DTranspose(filters, 3, padding="same")(x)
        x = layers.BatchNormalization()(x)
        x = layers.UpSampling2D(2)(x)

        # Project residual
        residual = layers.UpSampling2D(2)(previous_block_activation)
        residual = layers.Conv2D(filters, 1, padding="same")(residual)
        x = layers.add([x, residual])
        previous_block_activation = x  # Set aside next residual

    outputs = layers.Conv2D(num_classes, 3, activation="softmax", padding="same")(x)
    model = keras.Model(inputs, outputs)
    return model

# Get the model
num_classes = 2
model = get_unet_model(img_size, num_classes)
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(images, masks, test_size=0.2, random_state=42)

# Define callbacks with increased patience for early stopping
callbacks = [
    keras.callbacks.ModelCheckpoint("unet_model.h5", save_best_only=True),
    keras.callbacks.EarlyStopping(patience=10, monitor='val_loss')  # Increased patience
]

# Train the model
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=8,
    validation_data=(X_val, y_val),
    callbacks=callbacks
)
```

```python
# Load best model
model = keras.models.load_model("unet_model.h5")

# Evaluate on validation set
val_loss, val_accuracy = model.evaluate(X_val, y_val)
print(f"Validation accuracy: {val_accuracy:.4f}")

# Predict on validation set
preds = model.predict(X_val)

# Plot some predictions
def display_predictions(images, masks, preds, num=5):
    plt.figure(figsize=(20, 10))
    for i in range(num):
        plt.subplot(3, num, i + 1)
        plt.imshow(images[i])
        plt.axis('off')
        plt.subplot(3, num, i + 1 + num)
        plt.imshow(masks[i].squeeze(), cmap='gray')
        plt.axis('off')
        plt.subplot(3, num, i + 1 + num * 2)
        plt.imshow(np.argmax(preds[i], axis=-1), cmap='gray')
        plt.axis('off')
    plt.show()

display_predictions(X_val, y_val, preds)
```
